```python
!pip install -q gtts ipywidgets
import IPython.display as ipd
import ipywidgets as widgets
import random
from gtts import gTTS
import time
def generate_audio_files(pairs):
    for pair in pairs:
        for word in pair:
            try:
                tts = gTTS(word, lang='de')
                tts.save(f"{word}.mp3")
            except:
                print(pair)

minimal_pairs = [
    ("Hassen", "Hasen"),
    ("Tode", "Tote"),
    ("Miete", "Mitte"),
    ("Apfel" , "Äpfel"),
    ("Hohe" , "Höhe"),
    ("Biete","Bitte"),
    ("Geist","Gest"),
    ("Bett","Brett"),
    ("Kiefer","Käfer"),
    ("Wand","Wend"),
    ("Höhle","Hölle"),
    ("Hüte","Hütte"),
    ("Beet", "Bett"),
    ("Gasse","Kasse"),
    ("Grippe","Krippe"),
    ("Barbier","Papier"),
    ("Kirche","Kirsche"),
    ("Kirche","Küche"),
    ("Küche","Kirsche"),
    ("Kiste","Küste"),
    ("Ofen","Offen"),
    ("Schräg","Schreck"),
    ("Tier", "Tür"),
    ("Ohr","Uhr"),
    ("spuken","spucken"),
    ("drücken","drucken"),
    ("Bar","Bär"),
    ("zahlen","zählen"),
    ("schon","schön"),
    ("Bär","Bier"),
    ("Wall", "Wal"),
    ("fuhr","für"),
    ("Pass", "Bass"),
    ("Oper", "Ober"),
    ("Paar", "Bar"),
    ("tanken", "danken"),
    ("Tier", "dir"),
    ("baten", "baden"),
    ("Karten", "Garten"),
    ("Kunst", "Gunst"),
    ("Ecke", "Egge"),
    ("im Haus", "im Aus"),
    ("aus Hessen", "aus Essen"),
    ("wie hier", "wie ihr"),
    ("fischen", "wischen"),
    ("vier", "wir"),
    ("fangen", "Wangen"),
    ("heißer", "heiser"),
    ("reißen", "reisen"),
    ("fließen", "Fliesen"),
    ("Wicht", "wischt"),
    ("Kirsche", "Kirche"),
    ("Menschen", "Männchen"),
    ("mein", "nein"),
    ("zähmen", "Zähnen"),
    ("am", "an"),
    ("dünner", "Dünger"),
    ("ran", "Rang"),
    ("Land", "langt"),
    ("dümmer", "Dünger"),
    ("Lamm", "lang"),
    ("schlimmen", "Schlingen"),
        ("beten", "Betten"),
    ("fehle", "felle"),
    ("Weg", "weg"),
    ("stehlt", "stellt"),
    ("Reeder", "Retter"),
    ("gähnt", "Gent"),
    ("bieten", "bitten"),
    ("vieler", "Filter"),
    ("Wiege", "Wicke"),
    ("stiehlt", "stillt"),
    ("triet", "tritt"),
    ("ihn", "in"),
    ("Bahn", "Bann"),
    ("kam", "Kamm"),
    ("Staat", "Stadt"),
    ("Schlaf", "schlaff"),
    ("lahm", "Lamm"),
    ("Saat", "satt"),
    ("Schuster", "Schuss"),
    ("Mut", "Mutter"),
    ("Mus", "mus"),
    ("muß", "Stuhl"),
    ("le", "buk"),
    ("Buckel", "Bock"),
    ("Tone", "Tonne"),
    ("Ofen", "offen"),
    ("Moos", "Most"),  ("Hefe", "Höfe"),
    ("Lehne", "Löhne"),
    ("Sehne", "Söhne"),
    ("schon", "schön"),
    ("losen", "lösen"),
    ("Bogen", "Bögen"),
    ("Gent", "gönn"),
    ("helle", "Hölle"),
    ("kennen", "können"),
    ("konnte", "könnte"),
    ("Frosch", "Frösche"),
    ("Gott", "Götter"),
    ("Biene", "Bühne"),
    ("Kiel", "kühl"),
    ("liegen", "lügen"),
    ("Hut", "Hüte"),
    ("Blut", "Blüte"),
    ("Gut", "Güte"),
    ("Liste", "Lüste"),
    ("Kiste", "Küste"),
    ("Lifte", "Lüfte"),
    ("Kunst", "Künste"),
    ("Mutter", "Mütter"),
    ("Busch", "Büsche")
    # ... Add more pairs
]
generate_audio_files(minimal_pairs)
# Global variables to track the score, current pair, and the chosen word
score = 0
total_attempts = 0
current_pair = random.choice(minimal_pairs)
chosen_word = ""

current_audio = None
audio_output = widgets.Output()  # Create an Output widget for audio

def play_audio():
    global chosen_word, current_audio

    chosen_word = random.choice(current_pair)
    current_audio = ipd.Audio(f"{chosen_word}.mp3")
    with audio_output:
        ipd.clear_output(wait=True)
        display(current_audio)

def on_choice_made(b):
    global score, total_attempts, current_pair
    total_attempts += 1
    if b.description == chosen_word:
        b.button_style = 'success'
        score += 1
    else:
        b.button_style = 'danger'
    score_label.value = f"Score: {score}/{total_attempts}"

    # Wait briefly before moving to the next pair
    time.sleep(1)
    current_pair = random.choice(minimal_pairs)
    update_buttons()
    play_audio()  # Automatically play audio for the new word

def update_buttons():
    button1.description = current_pair[0]
    button1.button_style = ''
    button2.description = current_pair[1]
    button2.button_style = ''

# UI Components
button1 = widgets.Button(description=current_pair[0])
button2 = widgets.Button(description=current_pair[1])
score_label = widgets.Label(value=f"Score: {score}/{total_attempts}")

# Event Handlers
button1.on_click(on_choice_made)
button2.on_click(on_choice_made)

# Display UI
display(widgets.HBox([button1, button2]))
display(score_label)
display(audio_output)  # Display the audio_output widget

# Initial Audio Play
play_audio()

```
